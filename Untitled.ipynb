```python
from requests_html import HTMLSession
import openpyxl
import os
import time

# Função para extrair dados do Amazon Vendor Central e fazer download
def extrair_dados_amazon(dado):
    session = HTMLSession()

    url = 'https://vendorcentral.amazon.com.br/retail-analytics/dashboard/real-time-sales'
    r = session.get(url)

    # Verifica se a página foi carregada corretamente
    if r.status_code != 200:
        print(f'Erro ao acessar {url}. Status code: {r.status_code}')
        return

    # Insere o dado no campo de pesquisa
    campo_asin = r.html.find('kat-input[kat-aria-label="Pesquisa de ASIN"]', first=True)
    campo_asin.send_keys(dado)

    # Clica no botão 'Aplicar'
    btn_aplicar = r.html.find('kat-button[label="Aplicar"]', first=True)
    btn_aplicar.click()

    # Espera 5 segundos para o próximo clique
    time.sleep(5)

    # Clica no botão 'CSV'
    btn_csv = r.html.find('kat-button[label="CSV"]', first=True)
    btn_csv.click()

    # Espera 10 segundos para o download estar pronto
    time.sleep(10)

    # Clica no link 'Gerenciar downloads'
    link_gerenciar_downloads = r.html.find('a[id="downloadManager"]', first=True)
    link_gerenciar_downloads.click()

    # Espera 5 segundos para o próximo clique
    time.sleep(5)

    # Clica no link para fazer o download do arquivo
    link_download = r.html.find('a[href*="rads-ara-prod-na-us-east-1-downloaded-reports"]', first=True)
    download_url = link_download.attrs['href']

    # Realiza o download
    r_download = session.get(download_url)

    # Verifica se o download foi bem sucedido
    if r_download.status_code == 200:
        nome_arquivo = f'Inventario_{dado}.xlsx'
        diretorio_destino = r'C:\Users\lucas.oliveira\OneDrive - FRESNOMAQ IND DE MAQUINAS SA\4. INSIGHTS\Lucas\Base de dados\amazon asin'
        caminho_arquivo = os.path.join(diretorio_destino, nome_arquivo)

        with open(caminho_arquivo, 'wb') as arquivo:
            arquivo.write(r_download.content)

        print(f'Download para "{dado}" concluído e salvo em "{caminho_arquivo}".')

    else:
        print(f'Erro ao baixar dados para "{dado}". Status code: {r_download.status_code}')

    session.close()

# Função para ler dados do arquivo Excel
def ler_dados_excel(arquivo_excel):
    workbook = openpyxl.load_workbook(arquivo_excel)
    sheet = workbook.active

    dados = []
    for row in sheet.iter_rows(values_only=True):
        for cell in row:
            dados.append(cell)

    return dados

# Programa principal
if __name__ == '__main__':
    arquivo_excel = r'C:\Users\lucas.oliveira\OneDrive - FRESNOMAQ IND DE MAQUINAS SA\4. INSIGHTS\Lucas\Excel\asin.xlsx'

    dados_para_download = ler_dados_excel(arquivo_excel)

    for dado in dados_para_download:
        extrair_dados_amazon(dado)
        time.sleep(5)  # Aguarda 5 segundos entre cada download (ajuste conforme necessário)
```
